```python
# 한줄 모드(line mode) : % 한개
%timeit lst = [x**x for x in range(10000)]
```

```python
# 셀(여러 라인) 모드(cell mode) : %% 두개
%%timeit 
lst = []
for i in range(10000) :
    lst.append(i**i)
```
